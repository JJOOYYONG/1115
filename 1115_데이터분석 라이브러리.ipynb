```python
#%매직명령어
#그래픽 출력 과정을 볼 수 있도록 설정
%matplotlib inline

#라이브러리 들고오기
import numpy as np
import pandas as pd
import matplotlib
```

```python
#50개의 난수 생성
#넘파이: 데이터분석의 기본적인 기능들을 제공
#특히, 벡터 및 행령 연산과 관련된 편리한 기능들 제공
data=np.random.rand(50)
print(type(data))
print(data)
```

```python
#넘파이 배열을 판다스 시리즈 자료형 변환(인덱스,데이터의 조합)
#판다스 series,dataframe등의 자료구조를 활용하여 데이터 분석에 우수한 성능 발휘
#대량의 데이트를 빠른 속도로 처리가능

seri=pd.Series(data)
print(type(seri))
print(seri)
```

```python
seri.plot()
```

```python
#그래프 스타일 변경
matplotlib.style.use("ggplot")

#x축 인덱스 값, y축 랜던값

seri.plot()
```

```python
#rand(rows, cols)
#50개의 난수 생성
data_set=np.random.rand(10,3)
print(data_set)
print(type(data_set))
print(data_set.shape)
```

```python
#넘파이 행령을 판다스 데이터 프레임을 변환
#데이터프레임: 행과 열로 데이터를 조회할 수 있음
df=pd.DataFrame(data_set,columns=["A","B","C"])
df
```

```python
#데이터 프레임 자료를 그래프로 출력
df.plot()
```

```python
df.plot(kind="bar")
```

```python
df.plot(kind="barh")
```

```python
df.plot(kind="area")
```

```python
df.plot(kind="area",stacked=False)
```
